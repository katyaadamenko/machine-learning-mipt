Добавим в таблицу столбец time, в который запишем данной недели по порядку(за первую принята первая неделя 2012 года). 

```python
train["time"] = (train["year"] - 2012) * 52 + train["week"] + (train["year"] == 2015)
```

Отсортируем выборку по номеру товара, времени и сдвигу.

```python
train = train.sort(["item_id", "time", "shift"])
train
```

Можно предположить, что каждому товару сопоставлена некоторая большая строка признаков, каждой неделе соответстует массив признаков из строки с началом равным время - 52, а в таблице записан массив длиной 30, начало которого определяется временем и сдвигом (начало = время - 51 - сдвиг). Составим такие строки для каждого товара.

```python
train["data_shift"] = train["shift"]
train = train.drop(["shift"], 1)
train
```

```python
items = train["item_id"].unique()
```

```python
def find_item_frequencies(X):
    unique_keys = X["item_id"].unique()
    item_frequencies = {unique_key: 0 for unique_key in unique_keys}
    for i in xrange(X["item_id"].size):
        item_frequencies[X["item_id"][i]] += 1
    return item_frequencies
```

```python
train.index = xrange(len(train))
```

```python
train_item_frequencies = find_item_frequencies(train)
```

```python
train_item_frequencies
```

```python
items_features = {item:np.zeros(162 - 22) for item in items}
```

```python
for row in xrange(len(train)):
    beg = (train["time"][row] - train["data_shift"][row] - 51)
    items_features[train["item_id"][row]][beg : beg + 30] = np.array(train.iloc[row, 4 : 34])
```

```python
items_features
```

Обучим на полученных данных линейную регрессию.

```python
from sklearn.linear_model import LinearRegression
```

```python
model.coef_
```

Из получившихся коэффициентов видно, что самый важный - последний столбец.

```python
model.predict(X)
```

```python
def SMAPE(y_real, y_pred):
    return 100.0 / y_real.size * (np.abs(y_real - y_pred) / (np.abs(y_real) + np.abs(y_pred))).sum() * 2
```

```python
SMAPE(y, model.predict(X))
```
